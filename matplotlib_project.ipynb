```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = {'sayilar': np.random.randint(0,10001, size=1000)}
```

```python
df= pd.DataFrame(data)
```

```python
df
```

```python
df['durum'] = ['düşük' if x <=5000 else 'yüksek' for x in df['sayilar']]
```

```python
df
```

```python
def hesaplama(x):
    return x/2+5
```

```python
df['hesaplanan']= df['sayilar'].apply(hesaplama)
```

```python
df
```

```python
plt.plot(df['hesaplanan'], df['sayilar'])
plt.xlabel('Hesaplanan Sayılar')
plt.ylabel('Sayılarımız')
plt.title('Hesaplananlar ve sayılar')
plt.show()
```
